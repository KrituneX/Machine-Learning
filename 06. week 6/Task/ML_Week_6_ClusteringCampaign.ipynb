## Data preparation

### Subtask:
Prepare the data for clustering by standardizing numerical features and encoding categorical features.


**Reasoning**:
Standardize numerical features, encode categorical features, and combine them into a single DataFrame.



```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Feature Scaling
numerical_cols = df.select_dtypes(include=['number']).columns
numerical_cols = numerical_cols.drop('Z_CostContact')
numerical_cols = numerical_cols.drop('Z_Revenue')
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[numerical_cols])
scaled_df = pd.DataFrame(scaled_features, columns=numerical_cols)

# Categorical Feature Encoding
categorical_cols = ['Education', 'Marital_Status']
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded_features = encoder.fit_transform(df[categorical_cols])
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_cols))

# Combine Data
combined_df = pd.concat([scaled_df, encoded_df, df[['Z_CostContact', 'Z_Revenue']]], axis=1)
display(combined_df.head())
```

## Data clustering

### Subtask:
Apply the K-Means clustering algorithm to the prepared data. Determine the optimal number of clusters using the Elbow method.


**Reasoning**:
Apply the K-Means clustering algorithm and determine the optimal number of clusters using the Elbow method. Then train the final model with the optimal number of clusters.



```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia_values = []
cluster_range = range(2, 11)  # Test cluster numbers from 2 to 10

for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    kmeans.fit(combined_df)
    inertia_values.append(kmeans.inertia_)

# Plot the Elbow method graph
plt.figure(figsize=(8, 6))
plt.plot(cluster_range, inertia_values, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.show()

# Based on the Elbow method plot, choose the optimal number of clusters (e.g., 3)
optimal_k = 3  # Replace with the actual optimal k value observed from the plot

# Train the final KMeans model with the optimal k
final_kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
final_kmeans.fit(combined_df)

# Save the trained model and cluster labels
cluster_labels = final_kmeans.labels_
print(f"Optimal number of clusters: {optimal_k}")
print("Cluster Labels:", cluster_labels)
```

1. Silhouette Coefficient

```python
# Silhouette Coefficient
display(Math(r's(i) = \frac{b(i) - a(i)}{max(a(i), b(i))}'))
```

di mana:

s(i) adalah Silhouette Coefficient untuk data point i.
a(i) adalah rata-rata jarak antara data point i dan semua data point lainnya dalam cluster yang sama.
b(i) adalah rata-rata jarak antara data point i dan semua data point dalam cluster terdekat.
Silhouette Coefficient mengukur seberapa mirip data point dengan clusternya sendiri dibandingkan dengan cluster lain. Nilai mendekati 1 menunjukkan bahwa data point tersebut sangat mirip dengan clusternya sendiri dan tidak mirip dengan cluster lain. Nilai mendekati -1 menunjukkan bahwa data point tersebut lebih mirip dengan cluster lain daripada clusternya sendiri. Nilai sekitar 0 menunjukkan bahwa data point tersebut berada di antara dua cluster.

2. Calinski-Harabasz Index

```python
# Calinski-Harabasz Index
display(Math(r'CH = \frac{B(k)}{W(k)} \times \frac{n - k}{k - 1}'))
```

di mana:

CH adalah Calinski-Harabasz Index.
B(k) adalah dispersi antar-cluster (between-cluster dispersion).
W(k) adalah dispersi dalam cluster (within-cluster dispersion).
n adalah jumlah total data point.
k adalah jumlah cluster.
Calinski-Harabasz Index mengukur rasio dispersi antar-cluster dan dispersi dalam cluster. Nilai yang lebih tinggi menunjukkan bahwa cluster lebih terdefinisi dan terpisah, dengan dispersi antar-cluster yang besar dan dispersi dalam cluster yang kecil.

3. Davies-Bouldin Index

```python
# Davies-Bouldin Index
display(Math(r'DB = \frac{1}{k} \sum_{i}^{k} \max_{j \neq i} \left( \frac{s_i + s_j}{d(c_i, c_j)} \right)'))
```

di mana:

DB adalah Davies-Bouldin Index.
k adalah jumlah cluster.

d(cáµ¢, câ±¼) adalah jarak antara centroid cluster i dan cluster j.
Davies-Bouldin Index mengukur rata-rata kesamaan antara setiap cluster dan cluster yang paling mirip dengannya. Nilai yang lebih rendah menunjukkan cluster yang lebih baik, dengan jarak antar-cluster yang lebih besar dan dispersi dalam cluster yang lebih kecil.

4. Dunn Index

di mana:
